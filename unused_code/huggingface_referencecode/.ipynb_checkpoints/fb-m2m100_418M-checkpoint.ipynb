```python
!pip3 install torch torchvision torchaudio --extra-index-url https://download.pytorch.org/whel/cu113
!pip install sentencepiece

from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import sentencepiece

tokenizer = AutoTokenizer.from_pretrained("facebook/m2m100_418M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/m2m100_418M")
```

```python
generator = pipeline(task="text2text-generation", model=model, tokenizer=tokenizer)
```

```python
def generate_text(inp):
    output = generator(inp, forced_bos_token_id=generator.tokenizer.get_lang_id('en'))
    tln = output[0]
    for item in tln:
        result = tln[item]
    return result
```

```python
test = generate_text("Sea aqurium ada di pulau sentosa letaknya dekat universal studio namun jauh juga jalan nya y Awalnya masuk dan lihat sepertinya tidak ada yang istimewa seperti seaworld gitu namun masuk kedalam ketemu aquarium yang besarrr banget takjub deh banyak yang duduk2 disana modelnya kayak theatre Seperti akuarium pada umumnya ada show lumba2 juga dan ikan hiu besar2 Oh y tiket dewasa kalo beli go show 41 dolar cukup mahal ya mending beli di agen deh saya dapat cuma 260 ribu per orang Rekomen kalo ajak orang tua kemari karena di dalam dingin namun kuat jalan aja")
```

```python
text_col = pd.DataFrame(df, columns = ['Text'])

text_col
```

```python
text_col['english_text'] = text_col['Text'].apply(lambda x: generate_text(x))

text_col
```
