```python
import spacy
from spacy import displacy
import pandas as pd
```

```python
df=pd.read_excel('geo location 2021 tweets wo emoji translated.xlsx')  

df
```

```python
text = df["english_text"].to_string(index=False)
# text = str(df["english_text"])
text
```

```python
import spacy

nlp = spacy.load("xx_ent_wiki_sm")
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
displacy.render(doc, style="ent")
```
