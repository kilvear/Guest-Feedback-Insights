```python
import pandas as pd
```

```python
df = pd.read_excel("reviews_normalized_year.xlsx")
df
```

## Replace Klook ratings with 'Baik' to 8 and 'Sangat Direkomendasikan' to 10

```python
replace_rating = {'Baik': '8',
                  'Sangat Direkomendasikan': '10'}
```

```python
df = df.replace({'rating': replace_rating})
```

## Replace TripAdvisor ratings and normalize to 10

```python
mask = (df['source'] == 'tripadvisor')
df_tripadvisor = df[mask]
df_tripadvisor
```

```python
df.loc[mask, 'rating'] = (df.loc[mask, 'rating'].str[0].astype(int))*2
```

```python
df
```

## Replace Google Review ratings and normalize to 10

```python
mask2 = (df['source'] == 'google_reviews')
df_google = df[mask2]
df_google
```
